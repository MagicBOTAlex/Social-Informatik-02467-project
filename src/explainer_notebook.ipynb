2. Basic Statistics and Data Preprocessing

The dataset used in this project is based on the United Nations General Debate Corpus (UNGDC), which contains speeches delivered by member states during annual United Nations General Assembly debates.

The dataset covers the period from 1970 to 2015 and contains 7507 speeches from 199 unique countries.

The dataset was selected because it provides large-scale political communication data suitable for both text analysis and network analysis. Since countries frequently reference one another in speeches, the dataset naturally forms directional relationships that can be represented as a network.

The combination of textual speech data, country identifiers, and temporal information makes the dataset particularly useful for studying:
- international attention patterns
- political communication
- diplomatic relationships
- sentiment in international discourse

Before constructing the network, the textual data was cleaned and standardized in order to improve consistency when extracting country mentions and performing sentiment analysis.

Data Cleaning and Preprocessing

Before constructing the network, the text data was cleaned and prepared for analysis. This step was necessary, because political speeches contain variations in spelling, capitalization, punctuation and formatting.

The preprocessing process included standardizing the text, handling country names consistently and preparing the speeches for country mention extraction and sentiment analysis.

This ensured that references to countries could be identified more reliably across speeches and years.

3. Tools, Theory and Analysis

The project combines text analysis, sentiment analysis and network science in order to study relationships between countries in political speeches.

Each country is represented as a node in a directed network. When one of the countries mentions another country in a speech, a directed edge is created between them.

The direction of the edge is important, because political attention is not necessarily symmetrical. It is important to understand, that a mention from Denmark to China does not carry the same meaning as a mention from China to Denmark.

Edge weights represent the frequency of mentions between countries. In addition, sentiment analysis is used to estimate whether mentions are generally positive or negative.

Text Analysis and Country Mention Extraction

To identify the relationships between countries, the speeches were analyzed using text processing techniques. Country-names were extracted from speeches in order to identify, when one country referenced another country.

This process transforms textual political communication into a structured network representation where relationships between countries can be measured and analyzed quantitatively.

Text analysis techniques such as tokenization, text normalization and pattern matching were used to improve the reliability of country mention extraction across speeches.

Sentiment Analysis

In addition to identifying country mentions, sentiment analysis was used to estimate the tone of interactions between the countries.

This allows the project to capture not only how much attention countries give each other, but also whether the language used is generally more positive or negative.

By combining sentiment analysis with network analysis, the network becomes more informative than a traditional mention network, since edges can contain both frequency and sentiment information.

Network Analysis

After extracting country mentions, the relationships were modeled as a directed network where countries represent nodes and mentions represent edges.

Network analysis was used to study how political attention is distributed between countries and to identify important actors within the network.

Centrality measures were used to identify countries that receive large amounts of international attention or act as important connectors between countries.

The directed structure of the network also makes it possible to study asymmetrical relationships, where one country frequently references another without receiving the same level of attention in return.

4. Discussion

One limitation of the project is that the country mentions do not necessarily imply political agreement or disagreement. Diplomatic language is often complex, which can make the interpretation difficult.

Another limitation is that sentiment analysis models may struggle to fully capture the complexity of political communication, especially in formal diplomatic speeches where language can be indirect or ambiguous.

In addition, speeches show substantial variation in length and style, which may influence the number of country mentions and affect the structure of the network.

Despite these limitations, the project demonstrates how text analysis and network science can be combined to study international political communication in a large-scale and in a interactive way.
