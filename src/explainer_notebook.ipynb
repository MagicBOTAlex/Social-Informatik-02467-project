Explainer Notebook - Country Mentions in UN Speeches

This notebook explains the methodology, preprocessing, network construction and analysis behind our project on international country mentions in UN General Debate speeches.

1. Motivation

Our project explores how countries refer to each other in UN General Debate speeches. We were interested in understanding patterns of international attention and political communication through network analysis and sentiment analysis.

We chose this dataset because it contains rich political language and naturally forms relationships between countries through mentions in speeches.

The goal of the website is to allow users to explore global political attention patterns in an intuitive and interactive way.
